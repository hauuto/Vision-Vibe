```python
import os

img_folder = "img"
img_files = [f for f in os.listdir(img_folder) if f.lower().endswith(('.png', '.jpg'))]

fig, axes = plt.subplots(1, len(img_files), figsize=(5 * len(img_files), 10))
if len(img_files) == 1:
    axes = [axes]

for ax, fname in zip(axes, img_files):
    img_path = os.path.join(img_folder, fname)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    ax.imshow(img, cmap="gray")
    ax.set_title(fname)
    ax.axis("off")
    print(img_path)

plt.tight_layout()
plt.show()
```

```python
import os

img_folder = "img"
img_files = [f for f in os.listdir(img_folder) if f.lower().endswith(('.png', '.jpg'))]

fig, axes = plt.subplots(1, len(img_files), figsize=(5 * len(img_files), 10))
if len(img_files) == 1:
    axes = [axes]

for ax, fname in zip(axes, img_files):
    img_path = os.path.join(img_folder, fname)
    img = cv2.imread(img_path)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    ax.imshow(img_gray)
    ax.set_title(fname)
    ax.axis("off")
    print(img_path)

plt.tight_layout()
plt.show()
```
